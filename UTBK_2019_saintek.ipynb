```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
saintek = pd.read_csv('https://raw.githubusercontent.com/Peksyaji/UTBK-2019/main/score_science.csv')
saintek.head()
```

```python
saintek2 = saintek.drop(['Unnamed: 0','id_first_major','id_first_university','id_second_major','id_second_university','id_user'],axis=1)
saintek2.head()
```

```python
saintek2.info()
```

```python
saintek2.describe()
```

```python
# Define numeric columns on "Score Humanities" Data
num_score_saintek = ['score_bio', 'score_fis', 'score_kim', 'score_kmb','score_kpu'
                        , 'score_kua', 'score_mat', 'score_ppu']

# Create distribution plot on each columns
fig, ax = plt.subplots(4, 2, figsize = (14, 14))
fig.tight_layout(pad = 4)

for ax, n in zip(ax.flatten(), num_score_saintek):
    sns.distplot(ax = ax, a = saintek2[n].dropna(), label = "Skewness : %.2f"%(saintek2[n].skew()))
    ax.set_title(n, fontsize = 18)
    ax.legend(loc = 'best')

plt.show()
```

```python
ax = sns.boxplot(data=saintek2)
```

```python
# Create heatmap data numeric
cormat = saintek2.corr()
fig, ax = plt.subplots(figsize = (12, 8))
sns.heatmap(ax = ax, data = cormat, annot = True)
ax.set_yticklabels(cormat,rotation = 0)
plt.show()
```
